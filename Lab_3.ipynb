```python
import cv2
import numpy as np
# Install OpenCV library (run only once in Colab)
# Import OpenCV for image processing functions
# Import NumPy for numerical operations and arrays
import matplotlib.pyplot as plt   # Import matplotlib to display images
from google.colab import files   # Import files module to upload images in Colab
```

```python
uploaded = files.upload()

for fn in uploaded.keys():
    filename = fn

image = cv2.imread(filename)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
row, col, ch = image.shape

mean = 0
sigma = 25

gaussian = np.random.normal(mean, sigma, (row, col, ch))

noisy = image + gaussian
noisy = np.clip(noisy, 0, 255).astype(np.uint8)

plt.imshow(cv2.cvtColor(noisy, cv2.COLOR_BGR2RGB))
plt.axis("on")
plt.title("Gaussian Noisy Image")
plt.show()
```

```python
mean_filter = cv2.blur(noisy, (5,5))

plt.imshow(cv2.cvtColor(mean_filter, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Mean Filtered Image")
plt.show()
```

```python
gaussian_filter = cv2.GaussianBlur(noisy, (5,5), 0)

plt.imshow(cv2.cvtColor(gaussian_filter, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Gaussian Filtered Image")
plt.show()
```

```python
median_filter = cv2.medianBlur(noisy, 5)

plt.imshow(cv2.cvtColor(median_filter, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Median Filtered Image")
plt.show()
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

sharpened = cv2.filter2D(image, -1, kernel)

plt.imshow(cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Sharpened Image")
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(cv2.cvtColor(noisy, cv2.COLOR_BGR2RGB))
plt.title("Noisy")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(cv2.cvtColor(mean_filter, cv2.COLOR_BGR2RGB))
plt.title("Mean")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(cv2.cvtColor(gaussian_filter, cv2.COLOR_BGR2RGB))
plt.title("Gaussian")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(cv2.cvtColor(median_filter, cv2.COLOR_BGR2RGB))
plt.title("Median")
plt.axis("off")

plt.subplot(2,3,6)
plt.imshow(cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB))
plt.title("Sharpened")
plt.axis("off")

plt.show()
```
